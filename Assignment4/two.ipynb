```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('datasets_56097_107686_headbrain.csv')
data
```

```python
w0=[1,1,1,1,1,1.5,1.5,1.5,1.5,1,1,1,1,1,1]
w1=[1,1,1,1.5,1,1.5,1,1,1,1.5,1.5,1,3,1.5,1.5]
w2=[1,1.5,1,1,1,1,1.5,1,1,1.5,1,1.5,3,1.5,3]
w3=[1,1,1,1,1.5,1,1,1.5,1,1,1.5,1,3,3,1.5]
```

```python
e=[]
for i in range(0,15):
    ypre=[]
    for a in range(0,len(data['Gender'])):
        x1,x2,x3 = data['Gender'][a],data['Age Range'][a],data['Head Size(cm^3)'][a]
        ypre.append(w0[i]+w1[i]*x1+w2[i]*x2+w3[i]*x3)
    j=0
    for y in data['Brain Weight(grams)']:
        ypre[j]=y-ypre[j]
        ypre[j]=ypre[j]**2
        j+=1
    print(ypre)
    sum=0.0
    for y in ypre:
        sum+=y
    e.append(sum)
    
```

```python
print(e)
```

```python
index=[i+1 for i in range(0,15)]
x_index=np.array(index)
y_error=np.array(e)
```

```python
plt.plot(x_index,y_error)
```
